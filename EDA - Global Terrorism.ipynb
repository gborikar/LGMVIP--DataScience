# Plot graphs

```python
# Plot the number of terrorist attacks by year
plt.figure(figsize=(15, 10))
sns.countplot(x='iyear', data=df)
plt.xticks(rotation=90)
plt.title('Number of Terrorist Attacks by Year')
plt.show()
```

```python
# Plot the number of terrorist attacks by country
plt.figure(figsize=(15, 15))
sns.countplot(x='country_txt', data=df)
plt.xticks(rotation=90)
plt.title('Number of Terrorist Attacks by Country')
plt.show()
```

```python
df['country_txt'].value_counts().head(25)
```

```python
# Plot the number of terrorist attacks by TOP 25 countries
plt.figure(figsize=(15, 15))
sns.countplot(x='country_txt', data=df, order=df['country_txt'].value_counts().head(25).index)
plt.xticks(rotation=90)
plt.title('Number of Terrorist Attacks by Country (TOP 25)')
plt.show()
```

```python
# Plot the number of terrorist attacks by region
plt.figure(figsize=(15, 10))
sns.countplot(x='region_txt', data=df)
plt.xticks(rotation=90)
plt.title('Number of Terrorist Attacks by Region')
plt.show()
```

```python
df['nkill'].fillna(df['nkill'].mean(), inplace = True)
```

```python
# Plot the number of deaths by year
plt.figure(figsize=(15, 10))
sns.barplot(x='iyear', y='nkill', data=df, estimator=np.sum)
plt.xticks(rotation=90)
plt.title('Number of Deaths by Year')
plt.show()
```

```python
# Plot the number of deaths by country
plt.figure(figsize=(15, 10))
sns.barplot(x='country_txt', y='nkill', data=df, estimator=np.sum)
plt.xticks(rotation=90)
plt.title('Number of Deaths by Country')
plt.show()
```

```python
df['nkill'].value_counts().head(25).values
```

```python
# Plot the number of deaths by country for TOP 25 countries
plt.figure(figsize=(15, 10))
sns.barplot(x='country_txt', y='nkill', data=df.groupby('country_txt')['nkill'].sum().reset_index().sort_values(by='nkill', ascending=False).head(25))
plt.xticks(rotation=90)
plt.title('Number of Deaths by Country (TOP 25)')
plt.show()
```

```python
# Plot the number of deaths by region
plt.figure(figsize=(15, 10))
sns.barplot(x='region_txt', y='nkill', data=df, estimator=np.sum)
plt.xticks(rotation=90)
plt.title('Number of Deaths by Region')
plt.show()
```

```python
# Plot the number of deaths by weapons
plt.figure(figsize=(15, 10))
sns.countplot(x='weaptype1_txt',data=df ,order = df['weaptype1_txt'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Number of Deaths by Weapon Types')
plt.show()
```

```python
plt.figure(figsize=(15, 10))
sns.countplot(x='attacktype1_txt',data=df ,order = df['attacktype1_txt'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Number of Deaths by Attack Types')
plt.show()
```
